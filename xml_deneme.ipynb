```python
import xml.etree.ElementTree as ET

# Kullanıcıdan üst verileri toplama
def ust_veri_topla():
    return {
        "Baslik": input("Başlık girin: "),
        "Konu": input("Konu (Anahtar Kelimeler) girin: "),
        "Aciklama": input("Açıklama girin: "),
        "Olusturan": input("Oluşturan kişi/kuruluş girin: "),
        "Lisans": input("Lisans türünü girin: "),
        "Dil": input("Dil girin: "),
        "Kimlik": input("Kaynağın URL'sini girin: "),
    }

# Verileri XML formatında kaydetme
def xml_kaydet(ust_veri, dosya_adi="metadata.xml"):
    try:
        agac = ET.parse(dosya_adi)
        kok = agac.getroot()
    except FileNotFoundError:
        kok = ET.Element("AcikEgitimKaynaklari")

    kaynak = ET.SubElement(kok, "Kaynak")
    for anahtar, deger in ust_veri.items():
        ET.SubElement(kaynak, anahtar).text = deger

    with open(dosya_adi, "wb") as dosya:
        ET.ElementTree(kok).write(dosya, encoding="utf-8", xml_declaration=True)
    print(f"Veriler {dosya_adi} dosyasına kaydedildi.")

# XML dosyasından verileri okuma
def xml_oku(dosya_adi="metadata.xml"):
    try:
        agac = ET.parse(dosya_adi)
        kok = agac.getroot()
        for kaynak in kok.findall("Kaynak"):
            print("\nKaynak:")
            for eleman in kaynak:
                print(f"{eleman.tag}: {eleman.text}")
    except FileNotFoundError:
        print(f"{dosya_adi} dosyası bulunamadı.")


# XML dosyasından arama ve filtreleme
def xml_ara_filtrele(anahtar, deger, dosya_adi="metadata.xml"):
    try:
        agac = ET.parse(dosya_adi)
        kok = agac.getroot()
        bulunan = False
        for kaynak in kok.findall("Kaynak"):
            if kaynak.find(anahtar) is not None and kaynak.find(anahtar).text == deger:
                bulunan = True
                print("\nEşleşen Kaynak:")
                for eleman in kaynak:
                    print(f"{eleman.tag}: {eleman.text}")
        if not bulunan:
            print("Eşleşen kaynak bulunamadı.")
    except FileNotFoundError:
        print(f"{dosya_adi} dosyası bulunamadı.")

# Ana program
def ana_program():
    print("=== Açık Eğitim Kaynakları Üst Veri Toplayıcı ===")
    dosya_adi = "metadata.xml"
    while True:
        print("\n1. Üst Verileri Topla\n2. Üst Verileri XML'e Kaydet\n3. XML'den Verileri Oku\n4. Arama/Filtreleme Yap\n5. Çıkış")
        secim = input("Bir seçenek belirleyin: ")

        if secim == "1":
            ust_veri = ust_veri_topla()
        elif secim == "2":
            if 'ust_veri' in locals():
                xml_kaydet(ust_veri, dosya_adi)
            else:
                print("Henüz üst veri toplanmadı. Lütfen önce üst verileri toplayın.")
        elif secim == "3":
            xml_oku(dosya_adi)
        elif secim == "4":
            anahtar = input("Arama/filtreleme için bir anahtar girin (örn: Baslik, Konu, Dil): ")
            deger = input(f"{anahtar} için aranacak değeri girin: ")
            xml_ara_filtrele(anahtar, deger, dosya_adi)
        elif secim == "5":
            print("Programdan çıkılıyor.")
            break
        else:
            print("Geçersiz seçenek. Lütfen tekrar deneyin.")

if __name__ == "__main__":
    ana_program()

```
